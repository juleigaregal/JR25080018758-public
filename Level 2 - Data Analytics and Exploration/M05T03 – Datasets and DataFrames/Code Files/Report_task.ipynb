```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt',sep=' ')
df_balance = df
print(df_balance.groupby('Ethnicity')['Income'].mean())
print(df_balance.groupby('Married')['Balance'].mean())
print(df_balance['Income'].max())
print(df_balance['Income'].min())
print(df_balance['Cards'].sum())
print(df_balance.groupby('Gender')['Gender'].count())
```
