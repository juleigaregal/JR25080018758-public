```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('wine.csv')
print(df.head())
print(df.info())

df_wines_filtered = df[df['variety'].isin(['Cabernet Sauvignon', 'Pinot Noir','Chardonnay'])]
print(df_wines_filtered)


# Filter for the three wine varieties
df_wines_filtered = df[df['variety'].isin(['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay'])]

# Create a FacetGrid with histograms
fg = sns.FacetGrid(df_wines_filtered, col="variety", col_wrap=3, height=4, sharex=True, sharey=True)
fg.map_dataframe(sns.histplot, x="points", bins=10, kde=False, color="skyblue")

# Add titles and labels
fg.set_axis_labels("Points", "Count")
fg.set_titles(col_template="{col_name}")
plt.tight_layout()
plt.show()

```
