[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

```python
# Load data
import matplotlib.pyplot as plt
df = pd.read_csv("Cars93.csv")
```

### 1. Boxplot

```python
# Boxplot for revs per mile
filtered_df = df[df['Manufacturer'].isin(['Hyundai', 'Audi', 'Suzuki', 'Toyota'])]

# Make a boxplot for 'Revs per Mile' grouped by Manufacturer
plt.figure(figsize=(8, 5))
filtered_df.boxplot(column = 'Rev.per.mile', by='Manufacturer')

plt.title("Revs per Mile by Manufacturer")
plt.suptitle("")  # Removes the default 'Boxplot grouped by' title
plt.xlabel("Manufacturer")
plt.ylabel("Rev.per.mile")
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota


### 2. Histogram

```python
# Histogram of MPG
plt.figure(figsize=(8,5))
plt.hist(df['MPG.city'], bins=15, alpha=0.5, label='City MPG')
plt.hist(df['MPG.highway'], bins = 15, alpha = 0.5, label ='Highway MPG')

plt.title("city vs highway")
plt.suptitle(" ")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** 
It's more fuel efficient to drive on the highway, you get more mileage per gallon

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x = 'Wheelbase', y= 'Turn.circle', data = df)
plt.show()
plt.close()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  As the wheelbase increases, the turn cirlce does as well, so a positive correlation between them.



### 4. Barplot

```python
# Barplot for mean horsepower
mean_hp = df.groupby('Type')['Horsepower'].mean()
# Bar plot
plt.figure(figsize=(8, 5))
mean_hp.plot(kind='bar')

plt.title("Mean Horsepower by Car Type")
plt.xlabel("Car Type")
plt.ylabel("Mean Horsepower")
plt.xticks(rotation=45)
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes, it does appear to be the case
